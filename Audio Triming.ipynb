```python
from pydub import AudioSegment
import os
```

```python
# Path of input folder containing audio files
input_folder = "/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/Dataset/fakeaudio"

# Path of output folder where split audio files will be stored
output_folder = "/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/SpecDataset/FakeSpec"
```

```python
# Define function to split audio file into 3-second segments and save as WAV files
def split_and_save_audio(input_file_path, output_folder_path):
    # Load audio file
    audio = AudioSegment.from_file(input_file_path)
    
    # Define segment length in milliseconds
    segment_length_ms = 3000
    
    # Compute number of segments and duration of last segment
    num_segments = len(audio) // segment_length_ms
    last_segment_duration_ms = len(audio) % segment_length_ms
    
    # Split audio into segments of length segment_length_ms and save each segment as a WAV file
    for i in range(num_segments):
        start_time_ms = i * segment_length_ms
        end_time_ms = (i + 1) * segment_length_ms
        segment = audio[start_time_ms:end_time_ms]
        output_file_path = os.path.join(output_folder_path, os.path.splitext(os.path.basename(input_file_path))[0] + "_segment{}.wav".format(i))
        segment.export(output_file_path, format="wav")
        
    # Save last segment as a WAV file if it is longer than 100 milliseconds
    if last_segment_duration_ms > 100:
        start_time_ms = num_segments * segment_length_ms
        end_time_ms = start_time_ms + last_segment_duration_ms
        segment = audio[start_time_ms:end_time_ms]
        output_file_path = os.path.join(output_folder_path, os.path.splitext(os.path.basename(input_file_path))[0] + "_segment{}.wav".format(num_segments))
        segment.export(output_file_path, format="wav")
```

```python
# Iterate over audio files in input folder, split them into segments and save as WAV files in output folder
for filename in os.listdir(input_folder):
    if filename.endswith(".wav"):
        input_file_path = os.path.join(input_folder, filename)
        split_and_save_audio(input_file_path, output_folder)
```
