```python
import librosa
import os
import librosa.display
import numpy as np
import matplotlib.pyplot as plt
```

```python
# from google.colab import drive
# drive.mount('/content/drive')
```

```python
# Folder path containing the audio files
folder_path = '/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/Dataset/fakeaudio'
```

```python
# Folder path to save the STFT spectrograms
output_folder = '/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/Spectograms/FK'
```

```python
# Create the output folder if it doesn't exist
if not os.path.exists(output_folder):
    os.makedirs(output_folder)
```

```python
# Iterate through the files in the folder
for filename in os.listdir(folder_path):
    if filename.endswith('.wav'):
        # Load the audio file
        audio_file = os.path.join(folder_path, filename)
        audio, sr = librosa.load(audio_file)

        # Compute the STFT spectrogram
        stft = librosa.stft(audio)

        # Convert the magnitude spectrogram to dB scale
        spectrogram = librosa.amplitude_to_db(abs(stft))

        # Save the spectrogram as a PNG image
        output_filename = os.path.join(output_folder, filename[:-4] + '.png')
        plt.figure(figsize=(10, 4))
        librosa.display.specshow(spectrogram, sr=sr, x_axis=None, y_axis=None)
        plt.axis('off')
        plt.savefig(output_filename, bbox_inches='tight', pad_inches=0)
        plt.close()
```
