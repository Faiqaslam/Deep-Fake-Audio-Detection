```python
from pydub import AudioSegment
import os
import random
```

```python
# Define the input and output folder paths
input_folder = "/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/Dataset/fakeaudio/FakeAngryl"
output_folder = "/content/drive/MyDrive/Colab Notebooks/Deep Fake audio detection/Dataset/realaudio/Angry"
```

```python
# Loop through each audio file in the input folder
for filename in os.listdir(input_folder):
    # Check if the file is an audio file
    if filename.endswith(".wav") or filename.endswith(".mp3"):
        # Load the input audio file
        input_file = AudioSegment.from_file(os.path.join(input_folder, filename))

        # Define the number of segments to cut
        num_segments = random.randint(2, 3)

        # Define the length of each segment
        segment_length = len(input_file) // num_segments
         # Cut the audio into segments
        segments = []
        for i in range(num_segments):
            start = i * segment_length
            end = (i+1) * segment_length
            segments.append(input_file[start:end])
             # Shuffle the segments
        random.shuffle(segments)

        # Join the segments together to create the output audio file
        output_file = segments[0]
        for i in range(1, num_segments):
            output_file = output_file + segments[i]

        # Define the output file path
        output_path = os.path.join(output_folder, os.path.splitext(filename)[0] + "_output.wav")

        # Export the output audio file
        output_file.export(output_path, format="wav")
```
